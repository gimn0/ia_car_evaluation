```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier # biblioteca do knn
from sklearn.model_selection import train_test_split # validação cruzada
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gimn0/ia_car_evaluation/main/data/car.data', names=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'target']) # lê car.data, cria o dataframe e nomes nas colunas
df # mostra o dataframe
```

```python
df.info() 
```

```python
df.isnull().sum() # mostra quantos inputs nulos existem em cada coluna
```

```python
df.buying.replace( ('vhigh','high','med','low'), (1, 2, 3, 4), inplace=True )
df.maint.replace( ('vhigh','high','med','low'), (1, 2, 3, 4), inplace=True )
df.doors.replace( ('2','3','4','5more'), (1, 2, 3, 4), inplace=True )
df.persons.replace( ('2','4','more'), (1, 2, 3), inplace=True )
df.lug_boot.replace( ('small','med','big'), (1, 2, 3), inplace=True )
df.safety.replace( ('low','med','high'), (1, 2, 3), inplace=True )
df.target.replace( ('unacc','acc','good','vgood'), (1, 2, 3, 4), inplace=True )
# replace troca a tipagem objeto para int para que possamos aplicar a validação cruzada e determinar o knn score 
df
```

```python
data = df[['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety']] # entradas
target = df.target # saídas

data_train, data_test, target_train, target_test = train_test_split(data, target, test_size = 0.2, random_state = 50) # validação cruzada
```

```python
knn = KNeighborsClassifier(n_neighbors = 9) # 9 vizinhos é o que possibilita o melhor knn score
knn.fit(data_train, target_train)

print(knn.score(data_test, target_test))
```
